## Portfolio Project: GDP and Life Expectancy

The goal of this project is to analyze data on GDP (Gross Domestic Product) and life expectancy from the World Health Organization and the World Bank, identifying relationships between the GDP and life expectancy of six countries.

Investopedia.com defines GDP as: "Gross domestic product (GDP) is the total monetary or market value of all the finished goods and services produced within a country’s borders in a specific time period. As a broad measure of overall domestic production, it functions as a comprehensive scorecard of a given country’s economic health."

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
all_data = pd.read_csv('all_data.csv')

# Get a feel for the data
print(all_data.head())
print(all_data.info())
all_data.describe(include='all')
```

The earliest year is 2000, and the most recent year is 2015.

There are 96 rows, and it does not appear that there is any missing data.

```python
# Re-name the column: Life expectancy at birth (years) to one word to be in line with the other column names
all_data.rename(columns={'Life expectancy at birth (years)':'LEAB_Y'}, inplace=True)

# Check
print(all_data.head())
```

```python
# Which 6 countries are included in the data?
print(pd.unique(all_data.Country))
```

This list is a fair mix of first and third world countries.

We will first explore life expectancy:

Has life expectancy increased over time in the six nations?

```python
# Create scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(x='Year', y='LEAB_Y', hue='Country', data=all_data)
plt.ylabel('Life expectancy at birth (years)')
plt.title('Life Expectancy per Nation over Time')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.savefig('life_exp_per_nation_year.png')
plt.show()
plt.clf()
```

All the nations in our data set, excepct Zimbabwe, had a very slight increase in life expectancy over time.

Zimbabwe had a very low life expectancy at birth of about 46 years from the year 2000, and dropped slightly until 2004. However, from 2005 to 2016, there was a strong and steady increase in life expectancy to about 60 years of age.

Life expectancy in Zimbabwe is significantly lower than all the other nations in our data set.

Germany has the highest life expectancy, although Chile and the USA are close behind.

Next, let's see if GDP increased over time in the six nations?

```python
# Create lineplot
plt.figure(figsize=(8,6))
sns.lineplot(x='Year', y='GDP', hue='Country', data=all_data)
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.title('GDP per Nation over Time')
plt.savefig('gdp_per_nation_year_line.png')
plt.show()
plt.clf()
```

The USA and China's GDP have grown over time, with China having a sharp growth from about 2006. Mexico, Chile and Zimbabwe's GDP's have remained relatively flat and low, with Germany's GDP growing slightly, but tapering off quickly.

There is a significant difference in the GDP of the USA and the other countries. This may skew some of the data for the low GDP countries.
Let's plot each country's GDP on their own graph to see if we can see better trends.

```python
# Create dataframes for each specific country
all_data_chile=all_data[all_data['Country'] == 'Chile']
all_data_china=all_data[all_data['Country'] == 'China']
all_data_germany=all_data[all_data['Country'] == 'Germany']
all_data_mexico=all_data[all_data['Country'] == 'Mexico']
all_data_usa=all_data[all_data['Country'] == 'United States of America']
all_data_zim=all_data[all_data['Country'] == 'Zimbabwe']

# Create subplots
plt.figure(figsize=(10,6))
plt.subplot(2,3,1)
sns.lineplot(x='Year', y='GDP', data=all_data_chile)
plt.title('Chile')

plt.subplot(2,3,2)
sns.lineplot(x='Year', y='GDP', data=all_data_china, color='orange')
plt.title('China')

plt.subplot(2,3,3)
sns.lineplot(x='Year', y='GDP', data=all_data_germany, color='green')
plt.title('Germany')

plt.subplot(2,3,4)
sns.lineplot(x='Year', y='GDP', data=all_data_mexico, color='red')
plt.title('Mexico')

plt.subplot(2,3,5)
sns.lineplot(x='Year', y='GDP', data=all_data_usa, color='purple')
plt.title('USA')

plt.subplot(2,3,6)
sns.lineplot(x='Year', y='GDP', data=all_data_zim, color='brown')
plt.title('Zimbabwe')

plt.subplots_adjust(wspace=0.5)
plt.subplots_adjust(hspace=0.5)

plt.savefig('gdp_per_nation_year.png')
plt.show()
```

We can see better detail and trend in these individual graphs - all nations did see an overall growth in their GDP.

What is the average life expectancy & GDP in these nations?

```python
# Calculate mean GDP per country
all_data_mean_GDP = all_data.groupby('Country').GDP.mean().reset_index()
#print(all_data_mean_GDP)

# Calculate mean life expectancy per country
all_data_mean_LEAB = all_data.groupby('Country').LEAB_Y.mean().reset_index()
#print(all_data_mean_LEATB)

all_data_means=all_data_mean_GDP.merge(all_data_mean_LEAB)
print(all_data_means)
```

```python
# Bargraph showing average GDP in these nations
plt.figure()
sns.barplot(x='GDP', y='Country', data=all_data_means)
plt.title('Average GDP per Country')
plt.savefig('ave_gdp_per_country.png')
plt.show()
plt.clf()
```

```python
# Bargraph showing average life expectancy in these nations
plt.figure()
sns.barplot(x='LEAB_Y', y='Country', data=all_data_means)
plt.title('Average Life Expectancy at Birth per Country')
plt.savefig('ave_leab_per_country.png')
plt.show()
plt.clf()
```

Zimbabwe has the lowest average life expectancy at birth, of about 50 years. The rest of the countries have an average life expectancy of between 70 and 80 years.

Is there a correlation between GDP and life expectancy of a country?

```python
# Create subplots
plt.figure(figsize=(10,6))
plt.subplot(2,3,1)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_chile)
plt.title('Chile')

plt.subplot(2,3,2)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_china, color='orange')
plt.title('China')

plt.subplot(2,3,3)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_germany, color='green')
plt.title('Germany')

plt.subplot(2,3,4)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_mexico, color='red')
plt.title('Mexico')

plt.subplot(2,3,5)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_usa, color='purple')
plt.title('USA')

plt.subplot(2,3,6)
sns.scatterplot(x='LEAB_Y', y='GDP', data=all_data_zim, color='brown')
plt.title('Zimbabwe')

plt.subplots_adjust(wspace=0.5)
plt.subplots_adjust(hspace=0.5)

plt.savefig('leab_per_nation_year.png')
plt.show()
```

There appears to be a strong correlation between life expectancy and GDP for each country.

```python
# Import correlation
from scipy.stats import pearsonr

print("The correlation between each country's Life expectancy and it's GDP:")

corr_leab_gdp_chile, p = pearsonr(all_data_chile.LEAB_Y, all_data_chile.GDP)
print(f"Chile: {corr_leab_gdp_chile}")

corr_leab_gdp_china, p = pearsonr(all_data_china.LEAB_Y, all_data_china.GDP)
print(f"China: {corr_leab_gdp_china}")

corr_leab_gdp_germany, p = pearsonr(all_data_germany.LEAB_Y, all_data_germany.GDP)
print(f"Germany: {corr_leab_gdp_germany}")

corr_leab_gdp_mexico, p = pearsonr(all_data_mexico.LEAB_Y, all_data_mexico.GDP)
print(f"Mexico: {corr_leab_gdp_mexico}")

corr_leab_gdp_usa, p = pearsonr(all_data_usa.LEAB_Y, all_data_usa.GDP)
print(f"USA: {corr_leab_gdp_usa}")

corr_leab_gdp_zim, p = pearsonr(all_data_zim.LEAB_Y, all_data_zim.GDP)
print(f"Zimbabwe: {corr_leab_gdp_zim}")
```

These are very close to 1, meaning that GDP and life expectancy at birth is highly correlated.

What is the distribution of that life expectancy?

```python
# Create Histogram to determine the distribution of life expectancy
plt.figure()
sns.histplot(data=all_data, x='LEAB_Y', bins=50)
plt.title('Histogram of Life Expectancy at Birth')
plt.xlabel('Life Expectancy at Birth (years)')
plt.savefig('hist_leab_all.png')
plt.show()
plt.clf()
```

The histogram is left-skewed, showing that most of the countries in our data have higher life expectancies at birth. Zimbabwe had the lowest life expectancies, but they did increase.

## Conclusions

There is a very strong positive correlation between each country's life expectancy and it's GDP.

All countries did see an overall growth in their GDP over time, with China's GDP growing the sharpest.
